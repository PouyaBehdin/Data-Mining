```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Alter binary column variables to category types

df['Online']=df['Online'].astype('category')
df['CreditCard']=df['CreditCard'].astype('category')
df['Personal Loan']=df['Personal Loan'].astype('category')


# Practice binning numerical then categorical (won't be used in the naive bayes inputs or outputs anyways)

numerical = ['ID', 'Age', 'Experience', 'Income', 'ZIP Code', 'Family', 'CCAvg']

# Bin sizes

bin = {
    'ID': 1,
    'Age': 15,
    'Experience': 15,
    'Income': 50,
    'ZIP Code': 10000,
    'Family': 2,
    'CCAvg': 1,
}

# Binning

for col in numerical:
    bin_size = bin[col]
    df[col] = [round(val / bin_size) for val in df[col]]
    df[col] = df[col].astype('category')

# Splitting the data 60%/40%

predictors = ['Online', 'CreditCard']
outcome = ['Personal Loan']

X = pd.get_dummies(df[predictors], prefix_sep=' ', drop_first=False)
y = df[outcome]

train_df, valid_df = train_test_split(df, test_size=0.4, random_state=1)
print(train_df.shape, valid_df.shape)
```

```python
# 8.2.c.ii

import pandas as pd
from sklearn.model_selection import train_test_split

predictors = ['TRAF_CON_R', 'WRK_ZONE', 'LGTCON_I_R', 'WKDY_I_R', 'INT_HWY', 'WEATHER_R', 'SPD_LIM', 'TRAF_WAY', 'ALIGN_I', 'PROFIL_I_R', 'HOUR_I_R', 'SUR_COND']
outcome = ['INJURY']

X = pd.get_dummies(dff[predictors].astype('category'), prefix_sep='_', drop_first=False)
y = dff[outcome]['INJURY'].astype('category')  # Convert 'INJURY' to categorical data type

# Get the list of unique classes from the categorical variable y

classes = list(y.cat.categories)

# Split 60%/40%

X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.40, random_state=1)
print(X_train.shape, X_valid.shape)
```

```python
# run naive Bayes

acc2_nb = MultinomialNB(alpha=0.01)
acc2_nb.fit(X_train, y_train)

# predict probabilities

predProb_train = acc2_nb.predict_proba(X_train)
predProb_valid = acc2_nb.predict_proba(X_valid)

# predict class membership

y_train_pred = acc2_nb.predict(X_train)
y_valid_pred = acc2_nb.predict(X_valid)
```

```python
# Training Classification Matrix

classificationSummary(y_train, y_train_pred, class_names=classes)
```

```python
# Validation Classification Matrix

classificationSummary(y_valid, y_valid_pred, class_names=classes)
```
